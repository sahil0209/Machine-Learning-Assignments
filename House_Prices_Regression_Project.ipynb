## Preprocessing the categorical values

```python
categorical = df.select_dtypes(include=['object'])
categorical.head()
```

```python
# converting categorical values into dummies
cat_dummies = pd.get_dummies(categorical, drop_first=True)
cat_dummies.head()
```

```python
# drop categorical variables 
df = df.drop(list(categorical.columns), axis=1)


# concat dummy variables with train_house df
df = pd.concat([df, cat_dummies], axis=1)
df.head()
```

```python
df.drop('Id',axis=1,inplace=True)
```

## Making The Model

```python
y = df['SalePrice'].values
```

```python
df.drop('SalePrice',axis=1,inplace=True)
```

```python
X=df.values
```

```python
from sklearn.model_selection import train_test_split as tts
X_train,X_test,y_train,y_test = tts(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression as LR
from sklearn.neighbors import KNeighborsClassifier as KNN
from sklearn.svm import SVR
from sklearn.model_selection import cross_val_score as cvs
from sklearn.linear_model import LogisticRegression as logi
from sklearn.tree import DecisionTreeRegressor as dtr
from sklearn.preprocessing import StandardScaler as sc
from sklearn.ensemble import RandomForestRegressor as RFR
from sklearn.metrics import confusion_matrix as CM
from sklearn.metrics import accuracy_score as AS
from sklearn.metrics import r2_score
import sklearn.metrics as metrics
from sklearn.metrics import classification_report as CR
```

```python
sc_x = sc()
X_train = sc_x.fit_transform(X_train)
X_test = sc_x.transform(X_test)
```

```python
reg = [LR(),KNN(),SVR(kernel='linear'),logi(),dtr(),RFR()]
prediction_array = []
acc=[]
for i in reg:
    regressor=i
    regressor.fit(X_train,y_train)
    accuracy = cvs(estimator=regressor,X=X_train,y=y_train,cv=10)
    y_pred = regressor.predict(X_test)
    prediction_array.append(y_pred)
    acc.append(accuracy.mean())
    print(r2_score(y_test, y_pred))
```

```python
# Accuracy By Cross_Val_Score
acc
```

```python
# Now Since Random forest Regression has the highest accuracy we will go on to improve the models
ut = []
for i in range(1,101):
    g = RFR(n_estimators = i)
    g.fit(X_train,y_train)
    ut.append([g.score(X_train,y_train),i])
```

```python
print(max(ut))
```

```python
f = RFR(n_estimators = max(ut)[1])
f.fit(X_train,y_train)
m = cvs(estimator = g,X=X_train,y=y_train,cv=10)
print("Accuracy: ",m.mean())
print("Standard Deviation: ",m.std())
```
