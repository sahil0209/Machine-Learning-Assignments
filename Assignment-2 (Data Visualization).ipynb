The four features of the Iris dataset:

```python
import seaborn as sns
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
iris = sns.load_dataset('iris')
```

```python
iris.head()
```

```python
sns.pairplot(iris,hue='species')
```

Create a kde plot of sepal_length versus sepal width for setosa species of flower.

```python
s=iris[iris['species']=='setosa']
sns.kdeplot(data=s[['sepal_length','sepal_width']],cmap="plasma", shade=True, shade_lowest=False)
plt.title('Iris-setosa')
plt.xlabel('Sepal Length Cm')
plt.ylabel('Sepal Width Cm')
```

```python
tips=sns.load_dataset('tips')
tips.head()
```

```python
tips['tip_pct'] = 100 * tips['tip'] / tips['total_bill']
grid = sns.FacetGrid(tips, row="sex", col="time", margin_titles=True)
grid.map(plt.hist, "tip_pct", bins=np.linspace(0, 40, 15));
```

```python
# The code for the below given plot is:
sns.catplot(x="day", y="total_bill", hue="sex", kind="box", data=tips);
```

```python
# The Code for given plot is :
with sns.axes_style('white'):
    sns.jointplot("total_bill", "tip", tips, kind='hex')
```

```python
# The code for the below plot is :
sns.jointplot("total_bill", "tip", data=tips, kind='reg');
```

```python
# Code for plot given below
x=np.arange(0,10,0.1)
y=np.sin(x)
z=np.cos(x)
fig=plt.figure()
axes=fig.add_axes([0,0,1,1])
axes.plot(x,y)
axes.plot(x,z,linestyle='--')
```

```python
# Code for the Below 2 plots :
x=np.arange(0.1,10,0.1)
y=np.exp(x)
z=np.zeros(len(x))

fig,axes=plt.subplots(nrows=1,ncols=2,figsize=(10,5))
axes[0].plot(x,y)
axes[0].plot(x,z)
axes[1].set_yscale('log')
axes[1].plot(x,y)
axes[1].plot(x,x)
axes[0].set_title('Normal Scale')
axes[1].set_title('Logarithmic Scale')
```

```python
# Code for the plot below
fig=plt.figure()
ax=fig.add_axes([0,0,1,1])
x=np.arange(-1,1.1,0.01)
y=x**3
ax.plot(x,y)
ax.spines['left'].set_position(('data',0.0))
ax.spines['bottom'].set_position(('data',0.0))
ax.spines['top'].set_color('none')
ax.spines['right'].set_color('none')
```
